```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
basic = pd.read_excel('2021-2022basic.xlsx',index_col=0)
player_name = basic['Player']
player_name
plt.style.use('ggplot')
pd.set_option('display.max_columns', None) 
basic.round(4)
```

下面我们将加上筛选条件，画两幅图，其中第一幅图的筛选条件是：场均造成罚球的数量和场均命中罚球数，第二幅图的筛选条件是：斜率（罚球命中率）

```python
#在此我们筛选出场均罚球次数大于6个，场均命中罚球数量大于5个的全部球员，并显示他们是谁,以及他们的基本信息。
sns.scatterplot(x = 'FTA', y = 'FT', data = basic[(basic['FTA']>6) & (basic['FT']>5)], hue = 'Pos')
plt.legend(loc = 2, bbox_to_anchor = (1,1))
basic[(basic['FTA']>6) & (basic['FT']>5)][['Player','Pos','Tm','MP','3P%','2P%','FT','FTA','FT%','PTS']]
```

```python
#在这幅图中，我们筛选罚球总次数的同时还筛选罚球命中率，因为事实上可以发现好多人压根一场比赛也没多少罚球机会，样本太小。
#筛选掉小样本的方式是规定MP（minutes play）和G（game play)的最小值，这两个条件默认就放在我写的代码的这两个数字，但是还是给用户筛选的机会
sns.scatterplot(x = 'FTA', y = 'FT', data = basic[(basic['FT%']>0.9) & (basic['G']>30) & (basic['MP']>15)], hue = 'Pos')
plt.legend(loc = 2, bbox_to_anchor = (1,1))
basic[(basic['FT%']>0.9) & (basic['G']>30) & (basic['MP']>15)][['Player','Pos','Tm','MP','3P%','2P%','FT','FTA','FT%','PTS']]
```

上面分析了球员的场均罚球数，场均罚球命中数的关系，下面这部分一张图分析球员的罚球得分占比

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
total = pd.read_excel("2021-2022total.xlsx",index_col=0)
data1 = total[['Player','Pos','Age','Tm','G','MP','FG%','3P','3PA','3P%','2P','2PA','FT','FTA','PTS']]
data1 = data1[data1['PTS']>200]
data1['PTA'] = data1['PTS']/data1['G']
data1['MPA'] = data1['MP']/data1['G']
data1['FT%'] = data1['FT']/data1['FTA']
data1['%_ofPTSbyFT'] = data1['FT']/data1['PTS']
data1
```

```python
#罚球得分占总得分的比重,显然斜率越大代表越依赖罚球得分，这是一个总括，就放这就好了。
p1 = sns.scatterplot(x = 'PTS', y = 'FT', data = data1, hue = 'Pos')
p1.set_title("Free throw scores vs total scores")
plt.legend(loc = 2, bbox_to_anchor = (1,1))
```

```python
#根据上图进行筛选，筛选条件是罚球得分占比或者罚球得分数量,注意我这里写的是“或者”，即满足两个条件其中之一就可以
sns.scatterplot(x = 'PTS', y = 'FT', data = data1[(data1['%_ofPTSbyFT']>0.25) | (data1['FT']>400)], hue = 'Pos')
plt.legend(loc = 2, bbox_to_anchor = (1,1))
data1[(data1['%_ofPTSbyFT']>0.25) | (data1['FT']>400)][['Player','Pos','Tm','MPA','FT','FTA','FT%','PTS','PTA','%_ofPTSbyFT']]
```

```python
sns.scatterplot(x = 'PTS', y = '%_ofPTSbyFT', data = data1, hue = 'Pos')
plt.legend(loc = 2, bbox_to_anchor = (1,1))
plt.axhline(y = data1[data1['Pos']=='C']['%_ofPTSbyFT'].mean() ,ls = 'dotted',label = 'FT percent for C',color = 'r')
plt.axhline(y = data1[data1['Pos']=='PG']['%_ofPTSbyFT'].mean() ,ls = 'dashed',label = 'FT percent for PG',color = 'g')
plt.axhline(y = data1[data1['Pos']=='SG']['%_ofPTSbyFT'].mean() ,ls = 'solid',label = 'FT percent for SG',color = 'y')
plt.axhline(y = data1[data1['Pos']=='PF']['%_ofPTSbyFT'].mean() ,ls = 'dashdot',label = 'FT percent for PF',color = [0,0,1])
plt.axhline(y = data1[data1['Pos']=='SF']['%_ofPTSbyFT'].mean() ,ls = ':',label = 'FT percent for SF',color = [0,0.75,0.75])
plt.legend(loc = 2, bbox_to_anchor = (1,1))
```

通过上面这幅图我们可以看出，中锋的罚球得分占比最高，而得分后卫的发球得分占比最低。排二三名的是控球后卫和大前锋。
