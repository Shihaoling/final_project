```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import matplotlib.pyplot as plt

# Read and clean the data, Because the data is announced by the official, we do not consider the outliers.
regular = pd.read_csv('2021-2022 NBA Player Stats - Regular.csv',delimiter=";", encoding="latin-1", index_col=0)
```

```python
player=' '.join(regular['Player'].tolist())
NBA = np.array(Image.open('basketball.jpg'))
```

```python
from turtle import color


wordcloud = WordCloud(mask=NBA,background_color="white",max_words=100,colormap='Blues',width=100/120,height=100/120).generate(player)
fig, ax = plt.subplots(figsize=(3,3))
ax = wordcloud
plt.imshow(wordcloud, interpolation="bessel")
plt.axis("off")
plt.margins(x=0, y=0)
plt.savefig('D:\\PyCharm 2021.2\\MISY\\final_project\\1.png')
```
